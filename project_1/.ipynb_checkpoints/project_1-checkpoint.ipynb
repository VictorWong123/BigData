```python
stocks['Date'] = pd.to_datetime(stocks['Date'])
```

```python
stocks['Open'] = round(stocks['Open'],2)
```

```python
len(stocks['Symbol'].unique())
```

```python
from matplotlib.dates import YearLocator, DateFormatter

first_ten = stocks['Symbol'].unique()[:10]

for symbol in first_ten:  
    symbol_data = stocks[stocks['Symbol'] == symbol]
    plt.plot(symbol_data['Date'], symbol_data['Open'], label=symbol)



plt.xlabel('Date')
plt.ylabel('Open Price')


# Create a YearLocator to set ticks at yearly intervals
years = YearLocator()

# Set the locator for the x-axis ticks
plt.gca().xaxis.set_major_locator(years)

# Set the format for the x-axis tick labels to display the year only
plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))

plt.xticks(rotation=90)
plt.legend()
plt.show()
```
