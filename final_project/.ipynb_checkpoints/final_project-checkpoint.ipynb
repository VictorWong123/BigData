```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("StressLevelDataset.csv")
df_2 = pd.read_csv("Student Stress Factors.csv")
```

```python
df
```

```python
sum(df["stress_level"]>2)
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(df["anxiety_level"], df["headache"], c=df["stress_level"], cmap=sns.diverging_palette(250, 30, l=65, center="dark", as_cmap=True))
ax.set(title="Anxiety level and headaches",
       xlabel="Anxiety level",
       ylabel="headaches")

plt.legend(*scatter.legend_elements(),title="Stress Level",loc='center left', bbox_to_anchor=(1.0, 0.5));


```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(df["depression"], df["future_career_concerns"], c=df["stress_level"], cmap=sns.diverging_palette(250, 30, l=65, center="dark", as_cmap=True))
ax.set(title="Depression and future career concerns",
       xlabel="Depression",
       ylabel="Future career concerns")

plt.legend(*scatter.legend_elements(),title="Stress Level",loc='center left', bbox_to_anchor=(1.0, 0.5));


```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(df["stress_level"], df["sleep_quality"], c=df["mental_health_history"], cmap="copper")
ax.set(title="quality sleep vs Anxiety level",
       xlabel="Anxiety level",
       ylabel="Sleep Quality")

plt.legend(*scatter.legend_elements(),title="Mental Health History",loc='center left', bbox_to_anchor=(1.0, 0.5));


```

```python
sns.color_palette()
```

```python
plt.figure(figsize=(20,20))

for i, column in enumerate(df.columns, 1):
    plt.subplot(7,3, i)
    sns.countplot(data=df, x=column, palette="mako")
    plt.title(column)
    
    max = df[column].max()
    ticks = range(0, max+1)
    labels = ticks
    
    if max <=5:
         plt.xticks(ticks, labels)
    else:
        plt.xticks(ticks, labels, rotation=30)


plt.suptitle("Frequency of each variable", fontsize=20, fontweight="bold",x=0.5,y=1);
plt.tight_layout()


plt.show()
plt.savefig('Frequency.png')
```

```python
correlation = df.corr()
correlation_stress = correlation['stress_level'].sort_values(ascending=False)
correlation_stress = correlation_stress.drop('stress_level')

plt.figure(figsize=(10, 5))
sns.heatmap(correlation_stress.to_frame(), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Stress Level')
plt.show()
```

```python
correlation = df.corr()
correlation_stress = correlation['depression'].sort_values(ascending=False)
correlation_stress = correlation_stress.drop('depression')

plt.figure(figsize=(10, 5))
sns.heatmap(correlation_stress.to_frame(), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Depression')
plt.show()
```

```python
correlation = df.corr()
correlation_stress = correlation['academic_performance'].sort_values(ascending=False)
correlation_stress = correlation_stress.drop('academic_performance')

plt.figure(figsize=(10, 5))
sns.heatmap(correlation_stress.to_frame(), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with academic performance')
plt.show()
```
