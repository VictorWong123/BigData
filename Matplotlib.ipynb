```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
fig, (a1,a2) = plt.subplots(nrows=1,
                          ncols=2,
                      figsize=(10,5))

a1.set(title="random graph",
       xlabel ="years",
       ylabel="money"
      )


a1.plot(x,y)
a2.scatter(x,y);
```

```python
fig, ((a1,a2), (a3,a4)) = plt.subplots(nrows=2,
                          ncols=2,
                      figsize=(10,5))

a1.set(title="random graph",
       xlabel ="years",
       ylabel="money"
      )

a2.set(title="Another",
       xlabel ="years",
       ylabel="money"
      )

a3.set(title="something else",
       xlabel ="years",
       ylabel="money"
      )
a4.set(title="graph",
       xlabel ="years",
       ylabel="money"
      )

a1.plot(x,y)
a2.plot(x,y)
a3.plot(x,y)
a4.plot(x,y);
```
